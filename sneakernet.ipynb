```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        self.conv1 = nn.Conv2d(1, 6, 5)   ## conv layers 
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        
        self.fc1 = nn.Linear(7744, 120)   # linear layers 
        self.fc2 = nn.Linear(120, 84)
        self.out = nn.Linear(84, 1)     # final out channel

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        
        x = x.view(-1, 7744)  ## reshape thingy
        
        x = torch.relu(self.fc1(x))
        x = F.softmax(self.fc2(x), dim = 0)  # let this be softmax, relu exploded the loss 
        
        ##  loss isnt exploding
        
        x = self.out(x)
        return x


net = Net()
```
