```python
from PIL import Image
import os
import numpy
import matplotlib.pyplot as plt

## NIKE = 0

## PUMA = 1
```

```python

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        path = folder +"/"+ filename
        
        im = Image.open(path)
        images.append(im)
    return images
        


def convert_to_numpy(arr, vector):
    np_arr = []
    for p in range (len(arr)):
        mario =  [numpy.array(arr[p]), vector]
        
        np_arr.append(mario)
        
    return np_arr


```

```python
def check_samples(arr, gap):
    
    for l in range (0, len(arr), gap):
        
        plt.imshow(arr[l][0])
        plt.show()   ## shoving this in somehow shows all the images 
        
# plt.imshow(nike_final[0][0]) -- load up first image

# successfully loaded !
```

```python
def normalise(arr, width, height):
    arr_copy = arr
    for p in range (len(arr)):
        
        doofus = arr[p]
        
        doofus = doofus.resize((width, height), Image.ANTIALIAS)
        
        arr_copy[p]= doofus
        
    return arr_copy
        
       
def save(arr, name):
    fmt_name  = "numpy/" + name
    numpy.save(fmt_name, arr)
    
    print ("saved -- ", fmt_name)
```

```python
nike_img_arr = load_images_from_folder("nike_dump")
puma_img_arr = load_images_from_folder("puma_dump")


nike_norm = normalise(nike_img_arr , 100, 100)
puma_norm = normalise(puma_img_arr , 100, 100)


nike_final = convert_to_numpy(nike_norm, 0)
puma_final = convert_to_numpy(puma_norm, 1)


save(nike_final, "nike_final")
save(puma_final, "puma_final")
```

```python
check_samples(nike_final, 20)   # shows 5 samples each

check_samples(puma_final, 20)
```
